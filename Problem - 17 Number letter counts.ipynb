# 17 Number letter counts

<p>If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.</p>
<p>If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used? </p>
<br /><p class="note"><b>NOTE:</b> Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.</p>

```python
written_numbers = {
    "0" : "",
    "1" : "one",
    "2" : "two",
    "3" : "three",
    "4" : "four",
    "5" : "five",
    "6" : "six",
    "7" : "seven",
    "8" : "eight",
    "9" : "nine",
    
    "10" : "ten",
    "11" : "eleven",
    "12" : "twelve",
    "13" : "thirteen",
    "14" : "fourteen",
    "15" : "fifteen",
    "16" : "sixteen",
    "17" : "seventeen",
    "18" : "eighteen",
    "19" : "nineteen",
    
    "2x" : "twenty",
    "3x" : "thirty",
    "4x" : "forty",
    "5x" : "fifty",
    "6x" : "sixty",
    "7x" : "seventy",
    "8x" : "eighty",
    "9x" : "ninety",
    
    "xxx" : "hundred",
    
    "xxxx" : "thousand",
    
    "x" : "and"
}
```

```python
def write_number(number):
    number = str(number)
    size = len(number)
    
    if 0 <= int(number) < 20:
        return written_numbers[number]
    elif size == 2:
        if int(number) == 20:
            return written_numbers["2x"]
        if not int(number) % 10:
            return written_numbers[number[0]+"x"]
        return written_numbers[number[0]+"x"] + "-" + write_number(number[1])
    elif size == 3:
        if not int(number) % 100:
            return written_numbers[number[0]] + written_numbers["xxx"]
        return written_numbers[number[0]] + written_numbers["xxx"] + " and " + write_number(int(number[1:3]))
    elif size == 4:
        if number == "1000":
            return written_numbers[number[0]] + written_numbers["xxxx"]
        return written_numbers[number[0]] + written_numbers["xxxx"] + " " + write_number(number[1:4])
```

# 52 Permuted multiples


<p>It can be seen that the number, 125874, and its double, 251748, contain exactly the same digits, but in a different order.</p>
<p>Find the smallest positive integer, <i>x</i>, such that 2<i>x</i>, 3<i>x</i>, 4<i>x</i>, 5<i>x</i>, and 6<i>x</i>, contain the same digits.</p>


```python
def flip(x, i=-1, ii=-2):
    s = str(x)
    r = str()

    l = [a for a in s]
    l[i], l[ii] = l[ii], l[i]
    
    for b in l:
        r += b
        
    return r
```

```python
def perm(x):
    s = str(x)
    r = set()
    
    if len(s) == 1:
        r.add(s)
    elif len(s) == 2:
        r.add(s)
        r.add(flip(s))
    else:
        for a in range(1, len(s)+1):
            [r.add(flip(s, i=-len(s), ii=-a)[0]+x) for x in perm(flip(s, i=-len(s), ii=-a)[1:])]
    return r
```

```python
def check_mult(integer, *, factor=2):
    perms = perm(integer)
    check = (factor-1) * [False]
    
    for x in range(2, factor+1):
        if str(integer * x) in perms:
            check[x-2] = True
        else:
            break
            
    if False not in check:
        return True
    else:
        return False
```

# 49 Prime permutations

<p>The arithmetic sequence, 1487, 4817, 8147, in which each of the terms increases by 3330, is unusual in two ways: (i) each of the three terms are prime, and, (ii) each of the 4-digit numbers are permutations of one another.</p>
<p>There are no arithmetic sequences made up of three 1-, 2-, or 3-digit primes, exhibiting this property, but there is one other 4-digit increasing sequence.</p>
<p>What 12-digit number do you form by concatenating the three terms in this sequence?</p>

Anforderungen:
- alle Zahlen sind Primzahlen
- alle Zahlen sind vierstellig
- alle Zahlen sind eine Permutation
- alle Zahlen haben den gleichen Abstand (aufbauend)

```python
def fp(n):
    x = int(n**0.5)
    p = set()
    ps = {x: True for x in range(1, n+1)}
    ps[1] = False
    
    for X in range(2, x+1):
        y = 2*X
        while y <= n:
            ps[y] = False
            y += X
            
    psf = {key: value for (key, value) in ps.items() if value is not False}
    return list(psf.keys())
```

```python
def flip(x, i=-1, ii=-2):
    s = str(x)
    r = str()

    l = [a for a in s]
    l[i], l[ii] = l[ii], l[i]
    
    for b in l:
        r += b
        
    return r
```

```python
def perm(x):
    s = str(x)
    r = set()
    
    if len(s) == 1:
        r.add(s)
    elif len(s) == 2:
        r.add(s)
        r.add(flip(s))
    else:
        for a in range(1, len(s)+1):
            [r.add(flip(s, i=-len(s), ii=-a)[0]+x) for x in perm(flip(s, i=-len(s), ii=-a)[1:])]

    return r
```

```python
primes = [x for x in fp(10_000) if len(str(x)) == 4]
```

```python
len(primes)
```

```python
r = list()
for prime in primes:
    tr = set()
    permutations = perm(prime)
    
    if len(permutations) >= 3:
        for permutation in permutations:
            if int(permutation) in primes:
                tr.add(int(permutation))
                
    if len(tr) >= 3:
        r.append(tr)
```

```python
for ress in r:
    for value in get_diffs(ress).values():
        if value >= 3:
            print(get_diffs(ress), ress)
```

```python
def get_diffs(x):
    diffs = dict()
    s = sorted(list(x))
    
    for i, n in enumerate(s):
        for g in s[i+1:]:
            if g-n in diffs.keys():
                diffs[g-n][0] += 1
                diffs[g-n][2].append(g)
            else:
                diffs[g-n] = [1, g-n, [g]]
    
    return diffs
```

# 10 Summation of primes
<p>The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.</p>
<p>Find the sum of all the primes below two million.</p>

```python
import math
import time
```

```python
def is_prime(number):
    if number <= 1:
        return False
    for divisor in range(2, number):
        if not number % divisor:
            return False
    return True
```

```python
def sieve_primes_to(limit):
    candidates = [x for x in range(3, limit+1, 2)]
    candidates.insert(0, 2)
    calc_to = math.ceil(pow(limit, 0.5))
    
    for candidate in candidates[:calc_to]:
        if is_prime(candidate):
            for next_candidate in candidates[candidates.index(candidate):]:
                if not next_candidate % candidate and next_candidate != candidate:
                    candidates.remove(next_candidate)
    
    return candidates
```

```python
start = time.time()
p_2m = sieve_primes_to(2_000_000)
end = time.time()
print(sum(p_2m), ", calculated in ", round(end-start, 2), "s")
```

# Problem 1 - Multiples of 3 or 5

<p>If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.</p><p>Find the sum of all the multiples of 3 or 5 below 1000.</p>

```python
def find_multiples_sum(a, x):
    result = set()
    
    for number in a:
        for index in range(1, x):
            if not index % number:
                result.add(index)
    
    return sum(result)
```

```python
find_multiples_sum([3, 5], 10)
```

```python
find_multiples_sum([3, 5], 1000)
```

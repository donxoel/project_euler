<p>Starting with the number 1 and moving to the right in a clockwise direction a 5 by 5 spiral is formed as follows:</p>

<p>It can be verified that the sum of the numbers on the diagonals is 101.</p>
<p>What is the sum of the numbers on the diagonals in a 1001 by 1001 spiral formed in the same way?</p>



## Notes

We can get calculate the biggest number. Then, add every second number to 1. After adding 4 numbers expand the gap by 2 until biggest number was reached.

```python
def calc_diagonals(rowsize):
    result = [1]
    gap = 2
    trash = [0, None]
    
    for n in range(1, rowsize**2+1):
        if result[-1] + gap == n:
            result.append(n)
        
        if not (len(result)-1) % 4 and len(result)-1 not in trash:
            gap += 2
            trash[1] = len(result)-1
            if result[-1] + gap == n:
                result.append(n)
    
    return result
```

```python
sum(calc_diagonals(1001))
```

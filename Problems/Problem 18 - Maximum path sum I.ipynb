```python
triangle = [
                                [75],
                              [95, 64],
                            [17, 47, 82],
                          [18, 35, 87, 10],
                        [20,  4, 82, 47, 65],
                      [19,  1, 23, 75,  3, 34],
                    [88,  2, 77, 73,  7, 63, 67],
                  [99, 65,  4, 28,  6, 16, 70, 92],
                [41, 41, 26, 56, 83, 40, 80, 70, 33],
              [41, 48, 72, 33, 47, 32, 37, 16, 94, 29],
            [53, 71, 44, 65, 25, 43, 91, 52, 97, 51, 14],
          [70, 11, 33, 28, 77, 73, 17, 78, 39, 68, 17, 57],
        [91, 71, 52, 38, 17, 14, 91, 43, 58, 50, 27, 29, 48],
      [63, 66,  4, 68, 89, 53, 67, 30, 73, 16, 69, 87, 40, 31],
    [ 4, 62, 98, 27, 23,  9, 70, 98, 73, 93, 38, 53, 60,  4, 23]
]
```

pseudocode/ notation:

`nodes = [
    [(x, y), cost, cost_from_next(99-triangle[x][y]), previousnode]
]`

```python
unvisited_nodes, visited_nodes = list(), list()
```

```python
def init_dijkstra(triangle):
    a = -1

    for array in triangle:
        a += 1
        n = -1
        for number in array:
            n += 1
            unvisited_nodes.append([(a, n), 999_999, 99-triangle[a][n]])

    unvisited_nodes[0][2], unvisited_nodes[0][1] = 0, 0
```

```python
init_dijkstra(triangle)
```

```python
def get_neighbours(node):
    a = node[0][0] + 1
    b1 = node[0][1]
    b2 = node[0][1] + 1

    return get_node((a, b1)), get_node((a, b2))
```

```python
def get_node(tupleindices):
    a = tupleindices[0]
    b = tupleindices[1]
    nodelists = [visited_nodes, unvisited_nodes]
    
    for nodelist in nodelists:
        for node in nodelist:
            if node[0] == (a, b):
                return node
```

```python
def check_neighbours(previousnode, neighbournodes):
    if neighbournodes[0][1] > previousnode[1] + neighbournodes[0][2]:
        neighbournodes[0][1] = previousnode[1] + neighbournodes[0][2]
        neighbournodes[0].append(previousnode[0])
        
    if neighbournodes[1][1] > previousnode[1] + neighbournodes[1][2]:
        neighbournodes[1][1] = previousnode[1] + neighbournodes[1][2]
        neighbournodes[1].append(previousnode[0])
```

```python
def sort(node):
    return node[1]
```

```python
def finalsort(node):
    return [node[0][0], -node[1]]
```

```python
def get_shortest_path():
    path = list()
    end = [0, 999]
    prev = None
    
    for node in visited_nodes:
        if node[0][0] == 14:
            if node[1] < end[1]:
                end = node
                prev = node[3]
                
    path.append(end[0])
    
    while len(path) < 14:
        this = get_node(prev)
        path.append(this[0])
        prev = this[3]
    path.append(prev)
    
    return sorted(path)
```

```python
get_node((0, 0))
```

```python
def do_dijkstra():
    while 0 < len(unvisited_nodes):
        if unvisited_nodes[0][0][0] != 14:
            check_neighbours(unvisited_nodes[0], get_neighbours(unvisited_nodes[0]))
        visited_nodes.append(unvisited_nodes.pop(0))
        unvisited_nodes.sort(key=sort)
        
    visited_nodes.sort(key=finalsort, reverse=True)
```

```python
do_dijkstra()
```

```python
ssum = 0

for a, b in get_shortest_path():
    ssum += triangle[a][b]
    
ssum
```

<p>Find the value of <i>d</i> &lt; 1000 for which <sup>1</sup>/<sub><i>d</i></sub> contains the longest recurring cycle in its decimal fraction part.</p>

```python
def find_cycle(n):
    if len(str(n)) < 1000:
    
        digits = str(n).split(".")[1]
        cycle = digits[0]
        max_cycle = int()

        for i, digit in enumerate(digits):
            #print(digit, "==", cycle[0], ",", cycle, "==", digits[i:i+len(cycle)])
            if digit == cycle[0] and cycle == digits[i:i+len(cycle)]:
                if int(cycle) > int(max_cycle):
                    max_cycle = cycle
            else:
                if i == 0:
                    pass
                else:
                    cycle += digit
        #print("cycle :", max_cycle, " ", n)
        return max_cycle, n
    else:
        #print("no cycle: ", n)
        return 0, 1
```

```python
h = 0, 0
y = 0
for x in range(1,1001):
    if int(find_cycle(decimal.Decimal('1/x'))[0]) > int(h[0]):
        h = find_cycle(1/x)
        y = x
```

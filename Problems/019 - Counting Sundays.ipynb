```python
import math
```

```python
ws = {
    0 : "sunday",
    1 : "monday",
    2 : "tuesday",
    3 : "wednesday",
    4 : "thursday",
    5 : "friday",
    6 : "saturday"
}
```

```python
days = {
    "january" : 31,
    "february" : 28,
    "february_leap" : 29,
    "march" : 31,
    "april" : 30,
    "may" : 31,
    "june" : 30,
    "july" : 31,
    "august" : 31,
    "september" : 30,
    "october" : 31,    
    "november" : 30,
    "december" : 31
}
```

```python
def get_weekday(d, m, yyyy):
    if m <= 2:       # Jan/Feb → treat as months 11/12 of previous year
        m += 10
        yyyy -= 1
    else:
        m -= 2       # March = 1, April = 2, ..., December = 10
    y = yyyy % 100
    c = yyyy // 100
    w = (d + math.floor(2.6 * m - 0.2) - 2 + y + math.floor(y/4) + math.floor(c/4) - 2*c) % 7
    if w < 0:
        w += 7
    return ws[w], w
```

```python
get_weekday(10, 12, 1900)
```

```python
def check_sunday():
    count = 0
    for year in range(1901, 2001):
        for month in range(1, 13):
            _, w = get_weekday(1, month, year)
            if w == 0:  # Sunday
                count += 1
    return count
```

```python
def check_leap(year):
    if not year % 400:
        return True
    if not year % 4 and year % 100:
        return True
    return False
```

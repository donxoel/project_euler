# Problem 3 Largest Prime Factor

<p>The prime factors of $13195$ are $5, 7, 13$ and $29$.</p>
<p>What is the largest prime factor of the number $600851475143$?</p>

```python
def get_lpf(number):
    primes = get_primes(10_000)
    lpf = int()
    
    for prime in primes:
        if not number % prime:
            number /= prime
            lpf = prime
            
    if number == 1:
        return lpf
    else:
        return "We need more primes!"
```

```python
def get_primes(limit):
    number = 2
    primes = [2]
    
    while number < limit:
        prime_div_pass = True
        prime_to_be = int()
        
        for prime in primes:
            if not number % prime:
                prime_div_pass = False
        
        if prime_div_pass:
            primes.append(number)
            
        number += 1
            
    return primes
```

# 9 Special Pythagorean Triplet

<p>A Pythagorean triplet is a set of three natural numbers, $a \lt b \lt c$, for which,
$$a^2 + b^2 = c^2.$$</p>
<p>For example, $3^2 + 4^2 = 9 + 16 = 25 = 5^2$.</p>
<p>There exists exactly one Pythagorean triplet for which $a + b + c = 1000$.<br>Find the product $abc$.</p>

```python
pow(4, 2), pow(5, 2), pow(6, 2)
```

```python
def find_py_triplet(sum_abc):
    a, b, c, d = 1, 2, 3, round(sum_abc/2)
    
    for _a in range(1, d):
        for _b in range(_a, d):
            _c = pow(pow(_a, 2) + pow(_b, 2), 0.5)
            
            if not _c % 1 and (_a + _b + _c) == sum_abc:
                a, b, c, = _a, _b, round(_c)
    return a, b, c
```

```python
t = find_py_triplet(1000)
print(t)
t[0] * t[1] * t[2]
```

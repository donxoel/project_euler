# 15 Lattice paths

<p>Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.</p>

![Lattice paths](p015.png "Lattice paths")
<p>How many such routes are there through a 20×20 grid?</p>

## Notes

If the permutation word for a NE lattice path contains $n$ $N$ steps and $e$ $E$ steps, and if the path begins at the origin $(0, 0)$, then the path necessarily ends at $( e , n )$.

This means we need all possible permutations of 20 $E$s and 20 $N$s.

Also, the number of lattice paths from $( 0 , 0 )$ to $( n , k )$ is equal to the binomial coefficient ${n + k\choose k}$ or $\frac{(n+k)!}{k!(n-k)!}$.

```python
def factorial(number):
    result = 1
    
    while number > 1:
        result *= number
        number -= 1
        
    return result
```

```python
def binom_co(n, k):
    if n >= k:
        return factorial(n) / (factorial(k) * factorial((n-k)))
```

```python
def get_routes(n, k):
    return int(binom_co((n + k), k))
```

```python
get_routes(20, 20)
```

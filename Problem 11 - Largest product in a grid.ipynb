```python
def check_horizontally(array_of_arrays):
    gp = int()
    
    for array in array_of_arrays:
        for index in range(3, len(array)):
            product = array[index-3] * array[index-2] * array[index-1] * array[index]
            if product > gp:
                gp = product
    
    return gp
```

```python
check_horizontally(bro)
```

```python
def check_vertically(array_of_arrays):
    gp = int()
    
    for a in range(len(array_of_arrays[0])):
        for b in range(3, len(array_of_arrays)):
            product = array_of_arrays[b-3][a] * array_of_arrays[b-2][a] * array_of_arrays[b-1][a] * array_of_arrays[b][a]
            if product > gp:
                gp = product

    return gp
```

```python
check_vertically(bro)
```

```python
def check_diagonally(array_of_arrays):
    gp = int()
    
    #von links nach rechts
    for b in range(len(array_of_arrays)-3): # 0 bis 16
        for a in range(len(array_of_arrays[0])-3): # 0 bis 16
            product = array_of_arrays[b][a] * array_of_arrays[b+1][a+1] * array_of_arrays[b+2][a+2] * array_of_arrays[b+3][a+3]
            if product > gp:
                gp = product

            product = array_of_arrays[b][a+3] * array_of_arrays[b+1][a+2] * array_of_arrays[b+2][a+1] * array_of_arrays[b+3][a]
            if product > gp:
                gp = product

    return gp
```

```python
check_diagonally(bro)
```

```python
def check_largest_quad_product(array_of_arrays):
    largest_quad_products = [check_horizontally(array_of_arrays), check_vertically(array_of_arrays), check_diagonally(array_of_arrays)]
    
    return max(largest_quad_products)
```

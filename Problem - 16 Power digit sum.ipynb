# 16 Power digit sum

<p>2<sup>15</sup> = 32768 and the sum of its digits is 3 + 2 + 7 + 6 + 8 = 26.</p>
<p>What is the sum of the digits of the number 2<sup>1000</sup>?</p>



```python
def digit_sum(number):
    number = str(number)
    digit_sum = 0
    
    for digit in number:
        digit_sum += int(digit)
    
    return digit_sum
```

```python
digit_sum(32768)
```

```python
digit_sum(pow(2, 1000))
```
